```python
info_path="../../forced_lightcurves/sample_lc_v2/"
salt_df = pd.read_csv(info_path + "../../Nobs_cut_salt2_spec_subtype_pec.csv")
```

# Make Data Tables for the Paper

```python
res = pd.read_csv('results_40percent.csv')
res_tsquared = pd.read_csv('results_40_tsquared.csv')
```

```python
# Table with uninformative results
for tab in ['full', 'just15']:
    if tab == 'full':
        last_row = -1
        filename = 'uninformative_FULL.tex'
    elif tab == 'just15':
        last_row = 15
        filename = 'uninformative.tex'

    with open('../paper/tables/' + filename, 'w') as fw:
        print(r'\begin{deluxetable*}{lllccccccccccccccccc}', file=fw)
        print(r'\tabletypesize{\scriptsize}', file=fw)
        print(r'\tablewidth{0pt}', file=fw)
        print(r'\tablecaption{90\% Credible Regions for Marginalized Model Parameters (Uninformative Prior)\label{tab:uninformative}}', file=fw)
        print(r'\tablehead{', file=fw)
        print(r'\colhead{}', file=fw)
        print(r'& \colhead{}', file=fw)
        print(r'& \colhead{}', file=fw)
        print(r'& \multicolumn{3}{c}{$t_\mathrm{rise}$\,(d)}', file=fw)
        print(r'& \colhead{}', file=fw)
        print(r'& \multicolumn{3}{c}{$\alpha_g$}', file=fw)
        print(r'& \colhead{}', file=fw)
        print(r'& \multicolumn{3}{c}{$\alpha_r$}', file=fw)
        print(r'& \colhead{}', file=fw)
        print(r'& \multicolumn{3}{c}{$\alpha_r - \alpha_g$}', file=fw)
        print(r'& \colhead{}', file=fw)
        print(r'& \colhead{} \\', file=fw)
        # top line
        print(r'\cline{4-6}', file=fw)
        print(r'\cline{8-10}', file=fw)
        print(r'\cline{12-14}', file=fw)
        print(r'\cline{16-18}', file=fw)
        # center lines
        
        
        print(r'\colhead{ZTF Name}', file=fw)
        print(r'& \colhead{TNS Name}', file=fw)
        print(r'& \colhead{$z$\tablenotemark{\footnotesize{a}}}', file=fw)
        print(r'& \colhead{5}', file=fw)
        print(r'& \colhead{50}', file=fw)
        print(r'& \colhead{95}', file=fw)
        print(r'& \colhead{}', file=fw)
        print(r'& \colhead{5}', file=fw)
        print(r'& \colhead{50}', file=fw)
        print(r'& \colhead{95}', file=fw)
        print(r'& \colhead{}', file=fw)
        print(r'& \colhead{5}', file=fw)
        print(r'& \colhead{50}', file=fw)
        print(r'& \colhead{95}', file=fw)
        print(r'& \colhead{}', file=fw)
        print(r'& \colhead{5}', file=fw)
        print(r'& \colhead{50}', file=fw)
        print(r'& \colhead{95}', file=fw)
        print(r'& \colhead{reliable\tablenotemark{\footnotesize{b}}}', file=fw)
        print(r'& \colhead{normal\tablenotemark{\footnotesize{c}}}', file=fw)
    # 
#         print('\colhead{Name}', file=fw)
#         print('& \colhead{(J2000.0)}', file=fw)
#         print('& \colhead{(J2000.0)}', file=fw)
#         print('& \colhead{Name}', file=fw)
#         print('& \colhead{Name}', file=fw)
#         print('& \colhead{by}', file=fw)
#         print('& \colhead{(mag)}', file=fw)
#         print('& \colhead{Type}', file=fw)
#         print('& \colhead{}', file=fw)
#         print(r'& \colhead{(JD$^\prime$)\tablenotemark{d}}', file=fw)
#         print('& \colhead{(mag)}', file=fw)
#         print('& \colhead{(mag)}', file=fw)
#         print(r'& \colhead{(JD$^\prime$)\tablenotemark{d}}', file=fw)
#         print('& \colhead{(mag)}', file=fw)
#         print('& \colhead{(mag)}', file=fw)
        print('}', file=fw)
    #     print('\decimalcolnumbers', file=fw)
        print('\startdata', file=fw)
        for index, row in res.sort_values('ztf_name').iloc[0:last_row].iterrows():
            dat_string = ''
            ztf_name = str(row.ztf_name)[2:-1]
            dat_string += '{} & '.format(ztf_name)
            this_sn = np.where(salt_df.name == ztf_name)
            tns_name = salt_df['TNS name'].iloc[this_sn].values[0]
            dat_string += '{} & '.format(tns_name[0:2]+'\,'+tns_name[2:])
            if salt_df.z_source.iloc[this_sn].values[0] == 'Msl':
                dat_string += '{:.3f} & '.format(salt_df.z_adopt.iloc[this_sn].values[0])
            else:
                dat_string += '{:.4f} & '.format(salt_df.z_adopt.iloc[this_sn].values[0])
            dat_string += '{:.2f} & '.format(row.t_rise_05)
            dat_string += '{:.2f} & '.format(row.t_rise_50)
            dat_string += '{:.2f} & '.format(row.t_rise_95)
            dat_string += ' & '
            dat_string += '{:.2f} & '.format(row.flat_alpha_g_95)
            dat_string += '{:.2f} & '.format(row.flat_alpha_g_50)
            dat_string += '{:.2f} & '.format(row.flat_alpha_g_05)
            dat_string += ' & '
            dat_string += '{:.2f} & '.format(row.flat_alpha_r_95)
            dat_string += '{:.2f} & '.format(row.flat_alpha_r_50)
            dat_string += '{:.2f} & '.format(row.flat_alpha_r_05)
            dat_string += ' & '
            dat_string += '{:.2f} & '.format(row.flat_delta_95)
            dat_string += '{:.2f} & '.format(row.flat_delta_50)
            dat_string += '{:.2f} & '.format(row.flat_delta_05)
            if row.final_selection == 1:
                dat_string += r'y &'
            elif row.final_selection == 0:
                dat_string += r'n &'
            if (salt_df['Ia subtype'].iloc[this_sn].values[0] == 'normal' or
                salt_df['Ia subtype'].iloc[this_sn].values[0] == 'normal*' or
                salt_df['Ia subtype'].iloc[this_sn].values[0] == '99aa-like' or
                salt_df['Ia subtype'].iloc[this_sn].values[0] == '99aa-like*' or
                salt_df['Ia subtype'].iloc[this_sn].values[0] == '91T-like' or
                salt_df['Ia subtype'].iloc[this_sn].values[0] == '91T-like*'
               ):
                dat_string += r'y \\'
            else:
                dat_string += r'n \\'
                

            print(dat_string, file=fw)
        print(r'\enddata', file=fw)
        print(r'\tablecomments{', file=fw)

        print(r'''This table is available in its entirety in a machine-readable 
form in the online journal. A portion is shown here for guidance 
regarding its form and content. \\}''', file=fw)
        print(r'''The table includes the $5^\mathrm{th}$, $50^\mathrm{th}$, 
and $95^\mathrm{th}$ percentiles for the 4 parameters of interest: 
$t_\mathrm{rise}$, $\alpha_g$, $\alpha_r$, $\alpha_r - \alpha_g$. 
90\% credible regions are obtained by subtracting the $5^\mathrm{th}$ percentile 
from the $95^\mathrm{th}$ percentile. Estimates for $t_\mathrm{rise}$ come from 
$t_\mathrm{fl}$ while accounting for the uncertainties on $t_{B,\mathrm{max}}$, 
while estimates for the $\alpha$ parameters have been corrected to a flat prior
(see text for further details).
''', file=fw)
        print(r'''\tablenotetext{a}{Redshifts are reported to 4 decimal places, 
if the SN host galaxy redshift ($z_\mathrm{host}$) is known. 
Otherwise, the SN redshift ($z_\mathrm{SN}$) is reported to 3 decimal places 
(see \citealt{Yao19} for further details).}''', file=fw)
        print(r'''\tablenotetext{b}{Flag for SNe with reliable model parameters (see \ref{sec:qa}). 
y = reliable group, n = unreliable group (see text).}''', file=fw)
        print(r'''\tablenotetext{c}{Flag for normal SNe Ia,
y = normal, n = peculiar (the 7 peculiar SNe Ia in our sample are discussed in \ref{sec:rare}; their rise times are measured relative to $T_{g,\mathrm{max}}$).}''', file=fw)
        
        
        
#         print(r'\tablenotetext{c}{Determined from \texttt{SNID} (see text).}', file=fw)
#         print(r'\tablenotetext{d}{JD$^\prime$ = JD $-$ 2,458,000.}', file=fw)
        print(r'\end{deluxetable*}', file=fw)
```

```python
# Table with alpha=2 results
for tab in ['full', 'just15']:
    if tab == 'full':
        last_row = -1
        filename = 'alpha2_FULL.tex'
    elif tab == 'just15':
        last_row = 15
        filename = 'alpha2.tex'

    with open('../paper/tables/' + filename, 'w') as fw:
        print(r'\begin{deluxetable}{llcccc}', file=fw)
        print(r'\tabletypesize{\scriptsize}', file=fw)
        print(r'\tablewidth{0pt}', file=fw)
        print(r'\tablecaption{90\% Credible Region for $t_\mathrm{rise}$ ($\alpha = 2$ Prior)\label{tab:alpha2_rise}}', file=fw)
        print(r'\tablehead{', file=fw)
        print(r'\colhead{}', file=fw)
        print(r'& \colhead{}', file=fw)
        print(r'& \multicolumn{3}{c}{$t_\mathrm{rise}$\,(d)}', file=fw)
        print(r'& \colhead{} \\', file=fw)
        # top line
        print(r'\cline{3-5}', file=fw)
        # center lines
        
        
        print(r'\colhead{ZTF Name}', file=fw)
        print(r'& \colhead{TNS Name}', file=fw)
        print(r'& \colhead{5}', file=fw)
        print(r'& \colhead{50}', file=fw)
        print(r'& \colhead{95}', file=fw)
        print(r'& \colhead{reliable\tablenotemark{\footnotesize{a}}}', file=fw)
        print('}', file=fw)
        print('\startdata', file=fw)
        for index, row in res_tsquared.sort_values('ztf_name').iloc[0:last_row].iterrows():
            dat_string = ''
            ztf_name = str(row.ztf_name)[2:-1]
            dat_string += '{} & '.format(ztf_name)
            this_sn = np.where(salt_df.name == ztf_name)
            tns_name = salt_df['TNS name'].iloc[this_sn].values[0]
            dat_string += '{} & '.format(tns_name[0:2]+'\,'+tns_name[2:])
            dat_string += '{:.2f} & '.format(row.t_rise_05)
            dat_string += '{:.2f} & '.format(row.t_rise_50)
            dat_string += '{:.2f} & '.format(row.t_rise_95)
            if row.final_selection == 1:
                dat_string += r'y \\'
            elif row.final_selection == 0:
                dat_string += r'n \\'
                

            print(dat_string, file=fw)
        print(r'\enddata', file=fw)
        print(r'\tablecomments{', file=fw)

        print(r'''This table is available in its entirety in a machine-readable 
form in the online journal. A portion is shown here for guidance 
regarding its form and content. \\}''', file=fw)
        print(r'''The table includes the $5^\mathrm{th}$, $50^\mathrm{th}$, 
and $95^\mathrm{th}$ percentiles for  
$t_\mathrm{rise}$ after adopting the $\alpha = 2$ prior (see text for further details).
''', file=fw)
        print(r'''\tablenotetext{a}{Flag for SNe with reliable model parameters.
Note that the $\alpha = 2$ prior definition of reliable differs from that in 
\ref{sec:qa} (see text).}''', file=fw)
#         print(r'\tablenotetext{c}{Determined from \texttt{SNID} (see text).}', file=fw)
#         print(r'\tablenotetext{d}{JD$^\prime$ = JD $-$ 2,458,000.}', file=fw)
        print(r'\end{deluxetable}', file=fw)
```
